In this assignment I chose The Most Popular API from the New York Times, constructed an interface in pythin to read the JSON data, and transformed it into a pandas Dataframe.

```python
import requests
import pandas as pd
```

## The API Key

Your API key is entered below to connect to the New York Times API.

```python
api_key = input("Enter your API key: ")
```

## Getting the Data

I'll request data from the Most Popular API. This returns the most viewed articles from the last 7 days.

```python
url = "https://api.nytimes.com/svc/mostpopular/v2/viewed/7.json"

params = {
    "api-key": api_key
}

response = requests.get(url, params=params)
data = response.json()

data.keys()
```

## Converting the JSON to a DataFrame

The API response includes a "results" section, so I'll extract that and convert it into a pandas DataFrame.

```python
results = data["results"]
df = pd.json_normalize(results)

df.head()
```
