In this assignment, I collect ratings for six movies from several people that I know. Each person rates the movies they have seen on a scale from 1 to 5.

The goal is to analyze the ratings using pandas, calculate averages, and then compare the original ratings with normalized and standardized versions of the data.

```python
import pandas as pd
import numpy as np

ratings = pd.read_csv("movie_ratings.csv")

ratings
```

## Average Ratings

First, I calculate the average rating for each user and each movie based on the original ratings.

```python
# Average rating per user
user_avg = ratings.set_index("User").mean(axis=1)

user_avg
```

```python
# Average rating per movie
movie_avg = ratings.drop(columns="User").mean()

movie_avg
```

## Normalizing Ratings

Next, I create a new dataframe with normalized ratings for each user.

Normalization adjusts the ratings so that each user's ratings are centered relative to their own average rating. This helps remove personal rating bias from users who tend to rate movies higher or lower than others.

```python
ratings_numeric = ratings.set_index("User")

normalized_ratings = ratings_numeric.sub(ratings_numeric.mean(axis=1), axis=0)

normalized_ratings
```

## Average Ratings After Normalization

After normalizing the ratings, I calculate the averages again to compare them with the original values.

```python
# normalized averages by user
normalized_user_avg = normalized_ratings.mean(axis=1)

normalized_user_avg
```

```python
# normalized averages by movie
normalized_movie_avg = normalized_ratings.mean()

normalized_movie_avg
```

## Standardized Ratings

Standardization adjusts the ratings using both the mean and standard deviation. This makes it easier to compare ratings across users who may have very different rating habits.

```python
standardized_ratings = (ratings_numeric - ratings_numeric.mean(axis=1).values.reshape(-1,1)) / ratings_numeric.std(axis=1).values.reshape(-1,1)

standardized_ratings
```

## Average Ratings After Standardization

```python
standardized_ratings.mean(axis=1)
```

```python
standardized_ratings.mean()
```

## Conclusion

Using normalized ratings helps reduce differences in how people use the rating scale. Some users may tend to rate movies very generously while others rate more strictly. Normalization adjusts the ratings relative to each user's average, making comparisons between users more consistent.

However, normalization also removes some of the original meaning of the ratings because the scores no longer represent the actual rating scale from 1 to 5. This means the normalized values can be harder to interpret directly.

Standardized ratings go a step further by accounting for variation in ratings. This can make comparisons between users even more consistent, but it also makes the data more abstract compared to the original ratings.
