## Why I picked Amazon

Amazon is interesting to me because the recommendations are built right into the site. When I look at a product, I usually see things like “Frequently bought together” or “Customers who bought this item also bought.”

That makes it a good system to study because it is easy to see how the recommendations are being used.

## Scenario design

If I think about Amazon from the company’s side, the goal is probably to get people to buy more items and keep shopping on the site. So the recommendations are helping Amazon by making products easier to discover and maybe increasing sales.

If I think about it from the customer’s side, the goal is different. Customers want recommendations that are useful and relevant so they do not have to search as much. They want the site to help them find things they might actually want.

I think it makes sense to do scenario design twice because the company and the customer do not have the exact same goal, even though they are connected.

## What I can tell from the site

From the way Amazon shows recommendations, it looks like the site is using things like browsing history, purchase history, and similar products to suggest other items.

The recommendations on product pages make it seem like Amazon is trying to connect related items and show products that other people have also looked at or bought.

## My recommendations

One thing I would improve is making it clearer why a product is being recommended. A short explanation would make the system feel more helpful.

I also think it would be useful to give customers a little more control, like letting them choose to see more similar products or more different products.

Another possible improvement would be to make the recommendations a little less repetitive so the customer sees more variety.

## Conclusion

Overall, Amazon’s recommendation system seems like a strong example of a site that is trying to help both the company and the customer at the same time.

From what I can see, it is focused on showing products that are related to what the user is already looking at. That makes it useful, but I still think there is room to improve transparency and variety.
