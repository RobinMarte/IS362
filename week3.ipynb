```python

import pandas as pd
```

```python
cumulative_miles = pd.Series(
    data=[55, 120, 178, 241, 305],
    index=[1, 2, 3, 4, 5],    
    name="Cumulative Miles"
)

cumulative_miles
```

```python
daily_miles = cumulative_miles.diff()
daily_miles.iloc[0] = cumulative_miles.iloc[0]
daily_miles.name = "Daily Miles"

daily_miles
```

```python
summary = pd.DataFrame({
    "Cumulative Miles": cumulative_miles,
    "Daily Miles": daily_miles
})

summary
```
