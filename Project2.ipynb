## Dataset 1: Planetary Data

The dataset comes from NASA’s Jet Propulsion Laboratory planetary data table.

This table lists physical and orbital characteristics for the planets in our solar system, including measurements such as mass, gravity, orbital period, and distance from the sun.

The original dataset appears in a wide format where each planet is a row and different measurements are spread across multiple columns.

For this project, I created a CSV file that recreates this wide structure so that I can demonstrate how the data can be cleaned and prepared for analysis using pandas.

```python
import pandas as pd

planets = pd.read_csv("planets_wide.csv")

planets
```

### Issues With the Original Dataset

The original table contains several characteristics that make the dataset harder to analyze directly.

Some measurements use different units and multiple types of measurements are included in the same wide table. This means that before performing analysis it is helpful to isolate the variables that are relevant to the question being asked.

For this analysis I will focus on planetary mass and surface gravity.

```python
planet_analysis = planets[["Planet","Mass (10^24 kg)","Gravity (m/s^2)"]]

planet_analysis
```

```python
planet_analysis["Mass (10^24 kg)"].corr(planet_analysis["Gravity (m/s^2)"])
```

### Analysis

To explore the relationship between planetary mass and gravity, I calculated the correlation between the two variables.

The result shows a strong positive relationship between planetary mass and surface gravity. In general, planets with larger mass tend to have stronger gravity.

For example, Jupiter has the largest mass in the dataset and also one of the strongest gravity values. Smaller planets such as Mercury and Mars have much lower gravity values.

This result aligns with basic physics since gravitational force depends strongly on the mass of an object.

## Dataset 2: Video Game Sales

The second dataset comes from a Wikipedia page that lists the best-selling video games of all time.

The table contains several variables such as game title, release year, publisher, and sales figures across different regions including North America, Europe, Japan, and other regions.

The data appears in a wide format where regional sales are stored across multiple columns. This makes the dataset less flexible for analysis because the same type of variable (sales) is spread across several columns.

For this project, I recreated the table as a CSV file so that I can demonstrate tidying and transforming the data before performing analysis.

```python
import pandas as pd

games = pd.read_csv("video_games_wide.csv")

games
```

### Issues With the Original Dataset

The dataset stores sales values for different regions in separate columns. While this layout works well for displaying a table on a webpage, it makes analysis more difficult because the same variable (sales) appears across multiple columns.

A more tidy structure would store region as a variable and the corresponding sales values in a single column.

To address this, I reshape the dataset so that regional sales values are stored in a single column.

```python
tidy_games = games.melt(
    id_vars=["Title", "ReleaseYear", "Publisher"],
    value_vars=[
        "NA_Sales_Millions",
        "EU_Sales_Millions",
        "JP_Sales_Millions",
        "Other_Sales_Millions"
    ],
    var_name="Region",
    value_name="Sales_Millions"
)

tidy_games.head()
```

### Tidying the Data

Using the pandas melt() function converts the dataset from a wide format to a tidy format.

Instead of having multiple regional sales columns, the dataset now contains:

- a column that identifies the region
- a column that stores the sales value

This structure makes the dataset easier to analyze because all sales values are stored in one column.

```python
sales_by_year = games.groupby("ReleaseYear")["Global_Sales_Millions"].sum()

sales_by_year.sort_index()
```

### Analysis

Grouping the data by release year provides an overview of which years produced the highest-selling games in the dataset.

The results show that several modern games have very large global sales numbers, but some older titles also remain among the best-selling games of all time.

This suggests that while newer releases can achieve very high sales, classic games can continue to maintain strong long-term popularity.

## Dataset 3: New York City Climate Data

The third dataset comes from the Wikipedia page describing the climate of New York City.

The monthly climate table lists several measurements such as average high temperature, average low temperature, and precipitation.

The dataset is untidy because the months of the year are spread across multiple columns. In addition, different types of measurements are included in the same table.

For this project, I recreated the table as a CSV file in a similar wide format so that I can demonstrate how the data can be reshaped into a tidy structure before performing analysis.

```python
climate = pd.read_csv("nyc_climate_wide.csv")

climate
```

### Issues With the Dataset

In this table, the months (January through December) appear across multiple columns.

However, "month" is actually a variable and should appear in a single column rather than being spread across many columns.

In addition, the table mixes several different types of measurements together, including temperature and precipitation. This makes the dataset harder to analyze in its original structure.

To prepare the data for analysis, I reshape the dataset so that the month becomes a variable.

```python
climate_tidy = climate.melt(
    id_vars="Metric",
    var_name="Month",
    value_name="Value"
)

climate_tidy.head()
```

### Tidying the Data

Using the pandas melt() function converts the dataset from a wide format to a tidy format.

Instead of storing months across many columns, the dataset now includes:

- a Month column
- a Value column

Each row now represents a single observation for a specific metric and month. This structure makes it much easier to filter and analyze the data.

```python
high_temp = climate_tidy[climate_tidy["Metric"] == "AvgHighTemp_F"]
low_temp = climate_tidy[climate_tidy["Metric"] == "AvgLowTemp_F"]
precip = climate_tidy[climate_tidy["Metric"] == "Precipitation_in"]
```

## Analysis: Comparing Monthly Climate Values

After tidying the dataset, I can examine which months are the hottest, coldest, and wettest in New York City.

```python
high_temp.loc[high_temp["Value"].idxmax()]
```

```python
low_temp.loc[low_temp["Value"].idxmin()]
```

```python
precip.loc[precip["Value"].idxmax()]
```

### Analysis Results

The results show which months experience the highest temperatures, lowest temperatures, and the most precipitation in New York City.

Summer months appear as the warmest periods of the year, while winter months have the lowest temperatures. Precipitation levels vary throughout the year but can be identified clearly after restructuring the dataset.

Reshaping the dataset made this analysis possible because the month variable is now stored in a single column rather than being spread across multiple columns.
