```python
def checkNum(numbers):
    if numbers[0] == numbers[-1]:
        return True
    else:
        return False

numbers_x = [10, 20, 30, 40, 10]
numbers_y = [75, 65, 35, 75, 30]

print("Given list:", numbers_x)
print("result is", checkNum(numbers_x))

print("Given list:", numbers_y)
print("result is", checkNum(numbers_y))
```

```python
import pandas as pd

data = {
    "Student": ["Tommy", "Fred", "Gail", "User", "Randy", "Ariel"],
    "Hair Color": ["Brown", "Grey", "Blonde", "Brown", "Black", "Brown"],
    "Eye Color": ["Hazel", "Green", "Blue", "Brown", "Hazel", "Blue"],
    "Weight in kilograms": [60, 80, 76, 60, 88, 85]
}

df = pd.DataFrame(data)
df.index = [1, 2, 3, 4, 5, 6]

print(df)
print("\n")
print(df["Hair Color"])
print("\n")
print(df[["Eye Color"]])
print("\n")
print(df[["Student", "Weight in kilograms"]])
print("\n")
print(df.iloc[1:4])
```
