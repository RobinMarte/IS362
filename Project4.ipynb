In this notebook, I'll use the mushroom data directly from the UCI repository. I compare odor and habitat to see which one does a better job predicting whether a mushroom is poisonous.

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

## Loading the Data

I'll load only the columns I need from the Mushroom dataset.

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data"

mushrooms_raw = pd.read_csv(
    url,
    header=None,
    usecols=[0, 5, 22],
    names=["edible_poisonous", "odor", "habitat"],
    dtype=str
)

mushrooms_raw.head()
```

## Converting the Codes to Numeric Values

I'll replace the original codes with numeric values so I can use scikit-learn.

```python
mushrooms = pd.DataFrame({
    "edible_poisonous": mushrooms_raw["edible_poisonous"].replace({"e": 0, "p": 1}),
    "odor": mushrooms_raw["odor"].replace({
        "a": 0, "l": 1, "c": 2, "y": 3, "f": 4,
        "m": 5, "n": 6, "p": 7, "s": 8
    }),
    "habitat": mushrooms_raw["habitat"].replace({
        "g": 0, "l": 1, "m": 2, "p": 3, "u": 4, "w": 5, "d": 6
    })
})

mushrooms.head()
```

## Comparing the Two Predictors

I'll use scikit-learn to compare odor and habitat and see which one predicts poisonous mushrooms more accurately.

```python
y = mushrooms["edible_poisonous"]

X_odor = pd.get_dummies(mushrooms["odor"].astype(str), prefix="odor")
X_habitat = pd.get_dummies(mushrooms["habitat"].astype(str), prefix="habitat")

model = DecisionTreeClassifier(random_state=42)

odor_scores = cross_val_score(model, X_odor, y, cv=5)
habitat_scores = cross_val_score(model, X_habitat, y, cv=5)

print("Odor accuracy scores:", odor_scores)
print("Odor mean accuracy:", odor_scores.mean())

print("Habitat accuracy scores:", habitat_scores)
print("Habitat mean accuracy:", habitat_scores.mean())
```

## Conclusion

Based on the results, odor looks like the better predictor of whether a mushroom is poisonous.

Habitat does not look as strong, so odor seems like the more useful column for prediction.

If I were to keep going, I would test more columns together to see whether using multiple predictors improves the model.
